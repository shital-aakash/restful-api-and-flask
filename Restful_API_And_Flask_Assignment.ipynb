Que.1 What is a RESTful API?

   -> A RESTful API is a type of API that follows the principles of
      Representational State Transfer (REST), an architectural style for designing networked applications. It uses standard HTTP methods to interact with resources, making it flexible and scalable for various applications.

      Key Concepts:

      REST (Representational State Transfer):
      A set of constraints or guidelines for designing APIs that
      facilitate communication between systems.

      API (Application Programming Interface):
      A software interface that allows different applications to
      communicate and exchange data.

      HTTP Methods:
      REST APIs use standard HTTP methods like GET, POST, PUT, and DELETE
      to perform operations on resources.

      GET: Retrieves data.
      POST: Creates a new resource.
      PUT: Updates an existing resource.
      DELETE: Deletes a resource.

      Resources:
      Representations of data, often identified by a URL.

      State:
      The current state of a resource, which can be modified by the HTTP methods.

      Characteristics of a RESTful API:
      Stateless:
      Each request from a client to a server contains all the information necessary to process the request.

      Client-Server:
      The client (e.g., a web browser or mobile app) requests data or
      performs actions on the server, and the server responds.

      Cacheable:
      Responses from the server can be cached by clients for faster
      retrieval on subsequent requests.

      Uniform Interface:
      The API uses a consistent interface, making it easier to learn and use.

      Layered System:
      The client and server are independent layers, allowing for
      flexibility and scalability.

      Benefits of using RESTful APIs:

      Scalability: REST APIs are designed to handle a large number of
      requests efficiently.

      Flexibility: They can work with various programming languages and platforms.

      Simplicity: They are relatively easy to implement and use.
        
      Wide Adoption: REST APIs are widely used in web applications and
      mobile development.

      In essence, a RESTful API is a standardized and flexible way for different
      systems to interact over the internet, using the principles of REST to create
      a robust and scalable communication channe

Que.6 What are HTTP methods used in RESTful APIs?

   -> In RESTful APIs, the most commonly used HTTP methods are GET, POST, PUT,
      PATCH, and DELETE. These methods correspond to the CRUD (Create, Read, Update, Delete) operations, respectively. Other methods, like OPTIONS and HEAD, are less frequently used.

      Here's a more detailed explanation of each method:

      GET:
      Retrieves a representation of a resource from the server.
      It is used for reading data and should not be used for modifying resources.
        
      POST:
      Sends data to the server to create a new resource.
      It's often used for submitting form data or uploading files.

      PUT:
      Replaces an existing resource with the data provided in the request body.
      If the resource doesn't exist, it can create a new one.

      PATCH:
      Applies partial modifications to an existing resource.
      It updates specific fields rather than replacing the entire resource.

      DELETE:
      Removes a resource from the server.

Que.7 What is the purpose of the @app.route() decorator in Flask?

   -> The @app.route() decorator in Flask is used to bind a URL route to a
      specific function. It essentially tells Flask that when a user visits a particular URL, the function associated with that route should be executed. This mechanism is fundamental for creating web applications, as it allows different URLs to trigger different parts of the application's logic.

      When the Flask application receives a request, it checks the URL
      against the routes defined using @app.route().
      If a match is found, the corresponding function is called,
      and its return value is sent back to the client as the response.


Que.8 What is the difference between GET and POST HTTP methods?

   -> The primary difference between GET and POST HTTP methods is how they
      transmit data and their intended usage. GET requests retrieve data from a server by including parameters in the URL, while POST requests send data to the server within the body of the request. GET is for retrieving data (like displaying a webpage), while POST is for sending data to the server (like submitting a form).

      GET:
      Retrieves data from a server.
      Data is included as parameters in the URL (e.g., ?param1=value1¶m2=value2).
      Typically used for retrieving data that doesn't change the state
      of the server.
      Can be cached.
      May have character limits due to URL length restrictions.
      Less secure than POST because data is visible in the URL.

      POST:
      Sends data to a server to create, modify, or update resources.
      Data is included in the request body.
      Often used for submitting forms, uploading files, or sending data
      to the server.
      Cannot be cached.
      No character limits on the data being sent.
      More secure because data is not visible in the URL.

      In essence:
      GET is for getting data and POST is for posting (sending) data.

       app = Flask(__name__)

       @app.route('/data', methods=['POST'])
       def handle_post():
       if request.method == 'POST':
        data = request.get_json()  # Assuming the data is sent as JSON
        # Process the data (e.g., store it in a database)
        response = {'message': 'Data received successfully', 'data': data}
        return jsonify(response), 201  
        # Return a JSON response with status code 201 (Created)
    else:
        return 'Method not allowed', 405

    if __name__ == '__main__':
    app.run(debug=True)

      In this example, the /data route is set up to handle POST requests.
       When a POST request is sent to this route, the handle_post function
       is executed. The function retrieves the JSON data from the request
       body using request.get_json(), processes the data, and returns a
       JSON response with a success message and the received data.
       If any other type of request is sent to this route,
       it returns a "Method not allowed" error with a 405 status code.

       if __name__ == '__main__':

       it redirects to /target?param1=value1&param2=value2.

       query parameters param1 and param2 as arguments. The redirect() function
       then uses the generated URL to perform the redirection.

           Example Request:
           GET /search?q=flask → "Searching for: flask"

         🔹 3. Combining Both
            You can use both in the same route:
            @app.route('/user/<username>/posts')
            def user_posts(username):
            page = request.args.get('page', 1)  # Default to page 1
            return f"User: {username}, Page: {page}"

            Example Request:
            GET /user/alex/posts?page=2 → "User: alex, Page: 2"
